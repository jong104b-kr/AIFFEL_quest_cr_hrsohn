```python
# 제곱계산을 위한 math 모듈을 먼저 가지고 옵니다.
import math

# 첫 번째 정수, 연산자, 두 번째 정수를 차례대로 입력받습니다.
while True:
  print('지금부터 천하제일 계산대회를 시작한다.')
  try:
    su1 = int(input("정수 적어보소"))
  except ValueError: # 정수가 아닌 경우  ValueError를 처리하여 오류 메시지를 출력
    print("정수 뭔지 모르냐.")
    continue

  cal = input("+,-,/,*,** 중에 원하는 계산부호 적어보소:")
  if cal not in ['+','-','*','/','**']:
    print("없는거 넣지 말고야.")
    print("너 수학 안배웟냐아~~")
    continue

  try:
    su2 = int(input("두 번째 정수 적어보소"))
  except ValueError: # 정수가 아닌 경우  ValueError를 처리하여 오류 메시지를 출력
    print("정수야 정수라고야!!.")
    print("또 틀리믄 너는 14급 15급 공무원 해야돼")
    continue

  try:
    if cal == '+':
      result = su1 + su2
    elif cal =='-':
      result = su1 - su2
    elif cal =='*':
      result = su1 * su2
    elif cal =='/':
      result = su1 / su2
    elif cal == '**':
      result = su1 ** su2
    print(f"이거맞지?? {result}")
  except ZeroDivisionError:
    print("0으로 나눌순 없제~ 다시!.") # 숫자를 0으로 나누려고 할때 ZeroDivision Error를 통한 오류처리
    continue

#계산을 계속 이어갈지 입력받으며 continue or break로 처리
  continue_ok = input("계속 문제 내볼래?(yes, 그래, 좋아, y 외에는 기각이다.)")
  if continue_ok == "yes":
    continue
  elif continue_ok =="그래":
    continue
  elif continue_ok =="좋아":
    continue
  elif continue_ok =="y":
    continue

  else:
    print("계산 끝! 잘가고~~.")
    break

     #오류가 많이 발생했던 부분: indent오류
```

허재은 회고: 처음에 복잡해서 어떻게 시작할지 헤맸는데 순서대로 차례차례하니 해결된거 같다. indent오류가 자꾸 발생해서 주의해서 해야겠다.

손홍룡 회고 : 재은님 덕에 업혀갔다, 배운것이 많았다. 하나하나 하다보니 코딩을 바라보는 해상도가 조금 더 올라간 기분이다. 배움이 너무 딱딱하지 않도록 위트와 유머를 넣고 싶었다. 답답한 부분이 해소되지는 않지만 웃으면서 조금 더 감내할 수 있는 듯 하다. 앞으로도 즐겁게 코딩 공부를 해야겠다.
